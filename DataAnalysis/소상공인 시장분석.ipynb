## 동별로 편의점 개수 확인

```python
gu_list = df["행정동명"].value_counts().index.tolist()

for gu in gu_list:
    df_store = df[df["상권업종소분류명"] == "편의점"]
    print(gu, ": ", len(df_store[df_store["행정동명"] == gu].index))
```

```python
gu_list = df["행정동명"].value_counts().index.tolist()

for gu in gu_list:
    df_store = df[df["상권업종소분류명"] == "수퍼마켓"]
    print(gu, ": ", len(df_store[df_store["행정동명"] == gu].index))
```
