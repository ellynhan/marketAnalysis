```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../DATA/202_DT_B2_20210517105557.csv", encoding='cp949')
df.tail()
```

```python
dong_file = open("../DATA/부산_행정동.txt")
dong_list = {}
gu_list = []
curr_dong = ''
for dong in dong_file:
    dong = dong.strip()
    if dong.find('·') == -1:
        curr_dong = dong
        gu_list.append(curr_dong)
    else:
        for don in dong.split(' · '):
            dong_list[don] = curr_dong

print(dong_list)
```

```python
def makeDatesColumn():
    dates = []
    for year in range(2015, 2022):
        for month in range(1, 13):
            if year == 2021 and month > 3:
                continue
            dates.append("{0}. {1:0>2} 월".format(year, month))
    return dates

dates = makeDatesColumn()
```

```python
# for date in dates:
#     df[date] = df[]
print(df['구·군별'].value_counts().index.to_list())
```

```python
group_by_data = {}
for i in range(0, len(df.index), 8):
    gu = df["구·군별"][i]
    group_by_data[gu] = df.iloc[i: i + 8, :-1]
group_by_data['장전1동']
```

```python
final_data = pd.DataFrame(index = list(dong_list.keys()) + gu_list)
final_data['세대수'] = None
final_data['인구수'] = None
final_data
```

```python
from sklearn.linear_model import LinearRegression
predictDate = 202105
# 202105 예측결과
for dong_population in group_by_data:
    # print(dong_population)
    X = []
    y = []
    df = group_by_data[dong_population][(group_by_data[dong_population]['항목'] == "세대수") & (group_by_data[dong_population]['내외국인별'] == "총인구")].iloc[:, 4:]
    df = df.dropna()
    # print(df)
    # print(df.dtypes)
    if not df.empty:
        for col in df:
            if df[col].values[0] != '-':
                # print(int(col.split('. ')[0] + col.split(' ')[1]))
                X.append(int(col.split('. ')[0] + col.split(' ')[1]))
                # print(df[col].values[0])
                # print(df[col].values[0])
                y.append(int(df[col].values[0]))
    if X:
        X = np.array(X)
        y = np.array(y)
        reg = LinearRegression().fit(X.reshape(-1, 1), y)
        predict = reg.predict((np.array([[predictDate]])))[0]
        final_data['세대수'][dong_population] = predict
        print(dong_population, predict)
    else:
        print(dong_population, "NO")
```

```python
final_data["여성인구"] = None
final_data["외국인 수"] = None
predictDate = 202105
final_data
```

```python
# 202105 예측결과
for dong_population in group_by_data:
    X = []
    y = []
    df = group_by_data[dong_population][(group_by_data[dong_population]['항목'] == "여자인구") & (group_by_data[dong_population]['내외국인별'] == "총인구")].iloc[:, 4:]
    df = df.dropna()
    if not df.empty:
        for col in df:
            if df[col].values[0] != '-':
                X.append(int(col.split('. ')[0] + col.split(' ')[1]))
                y.append(int(df[col].values[0]))
    if X:
        X = np.array(X)
        y = np.array(y)
        reg = LinearRegression().fit(X.reshape(-1, 1), y)
        predict = reg.predict((np.array([[predictDate]])))[0]
        final_data['여성인구'][dong_population] = predict
        print(dong_population, predict)
    else:
        print(dong_population, "NO")
```

```python
# 202105 예측결과
for dong_population in group_by_data:
    X = []
    y = []
    df1 = group_by_data[dong_population][(group_by_data[dong_population]['항목'] == "인구") & (group_by_data[dong_population]['내외국인별'] == "총인구")].iloc[:, 4:]
    df2 = group_by_data[dong_population][(group_by_data[dong_population]['항목'] == "인구") & (group_by_data[dong_population]['내외국인별'] == "한국인")].iloc[:, 4:]
    df1 = df1.dropna()
    df2 = df2.dropna()
    if not df1.empty:
        for col in df1:
            if df1[col].values[0] != '-':
                X.append(int(col.split('. ')[0] + col.split(' ')[1]))
                y.append(int(df1[col].values[0]) - int(df2[col].values[0]))
    if X:
        X = np.array(X)
        y = np.array(y)
        reg = LinearRegression().fit(X.reshape(-1, 1), y)
        predict = reg.predict((np.array([[predictDate]])))[0]
        final_data['외국인 수'][dong_population] = predict
        print(dong_population, predict)
    else:
        print(dong_population, "NO")
```
